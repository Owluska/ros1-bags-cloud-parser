```python
import pandas as pd
import os
from math import floor
file = '/home/vist/scripts/throttle_const_20s.csv'
path, head = os.path.split(file)
name, ext = head.split(".")
data = pd.read_csv(file, index_col=0)
print(len(data['bag_name'].unique()))
data.head()
```

```python
data.columns
```

```python
first_columns = [
    'bag_name', 'time', 'vehid'
]

unique_columns = [
    'ego_status', 'ego_gear', 'throttle'
]

mean_columns = [
    'ego_speed', 'ego_x', 'ego_y', 'pitch'
]

min_columns = [
    'date'
]

max_columns = [
]

cols_dict = {
    'first': first_columns,
    'unique': unique_columns,
    'mean': mean_columns,
    'min': min_columns,
    'max': max_columns,
}

agg_dict = {
    col: func
    for func, cols in cols_dict.items()
    for col in cols
    if cols
}
# agg_dict
time_step = 60
data['time_rounded'] = data['time'].apply(lambda n: floor(n / time_step) * time_step)
data.groupby(by='bag_name', as_index=False).agg(agg_dict).to_csv(
    os.path.join(path, name + '_unique.' + ext))
```
